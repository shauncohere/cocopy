```python
import pandas as pd
```

```python
df=pd.read_csv("finetune.csv")
df.tail(10)
```

```python
data = df[df.Label==1]
```

```python
data.head(10)
```

```python
def write_rows_to_txt_file_with_seperator(df, train_filename, test_filename,seperator, split_size):
    with open(train_filename, "w") as f, open(test_filename, "w") as f2:
        for index,row in df.iterrows():
            if index % split_size == 0:
                f2.write(row["Example"] + "\n" + seperator + "\n")
            else:
                f.write(row["Example"] + "\n" + seperator + "\n")

write_rows_to_txt_file_with_seperator(data, "finetune.txt", "----", 10)

```
